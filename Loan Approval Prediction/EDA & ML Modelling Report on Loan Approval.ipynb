```python
# Dependents
# PLOT BETWEEN NUMBER OF DEPENDENTS AND MARITAL_STATUS
sns.countplot(hue='Dependents',x='Marital_Status',data=data)
plt.figure(figsize=(15,5))
plt.show()
```

##### As it is clearly visible that in case of not married the number of dependents is majorly 0 but in case of married the number of dependents takes other values as well. So extracting the highest among them that is 2 , Imputation has been made

```python
# In this function we are replacing the null values present in the DEPENDENTS columns :
# * If the marital_status is Yes then we are replacing the null values by 2 no of dependents
# * IF the marital_status is NO then we are relacing the null values by 0.
def impute_Dependents(cols):
    Dependents = cols[0]
    Marital_Status = cols[1]
    
    if pd.isnull(Dependents):

        if Marital_Status == 'Y':
            return ('2')

        else:
            return ('0')

    else:
        return (Dependents)
```

```python
data['Dependents'] = data[['Dependents','Marital_Status']].apply(impute_Dependents,axis=1)
data['Dependents'] = data['Dependents'].astype('int')

#TEST_DATA
test_data['Dependents'] = test_data[['Dependents','Marital_Status']].apply(impute_Dependents,axis=1)
test_data['Dependents'] = test_data['Dependents'].astype('int')
```

# TRAINING AND TESTING DATA

#### Let us train the given data with number of models and find out the efficiency of each and every model and choose the best

```python
# Impoerting train_test_split to seperate the training and testing data to fit it in the model for accuracy check and prediction
from sklearn.model_selection import train_test_split
# Following metrics will help us in evaluating the accuracy and various other parameters by the help of which we can decide which is a better model
from sklearn.metrics import classification_report,confusion_matrix
from sklearn.metrics import accuracy_score
x = data.drop(columns=['CPL_Status_Y'],axis=1) # This will contain the independent variables
y = data['CPL_Status_Y'] # This will contain the dependent variables
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.25, random_state=102) 
# Size of train data is 75%
# Size of test data is 25%
```

### 1. Logistic Regression

###### Logistic regression is a statistical model that in its basic form uses a logistic function to model a binary dependent variable, although many more complex extensions exist. In regression analysis, logistic regression (or logit regression) is estimating the parameters of a logistic model (a form of binary regression).


```python
from sklearn.linear_model import LogisticRegression
```

```python
model1 = LogisticRegression()
model1.fit(X_train,y_train) # Training the model on the train data
```

```python
prediction1 = model1.predict(X_test) # Predicting the values of the test data
```

```python
print(classification_report(y_test,prediction1)) # Report about the model performance
```

```python
accuracy_train = accuracy_score(y_test,prediction1)
print('accuracy_score on train dataset : ', accuracy_train) # Accuracy of the model on the basis of which we will choose our best model
```

```python
s1=round(model1.score(X_test,y_test)*100,3) # Converting the accuracy score into a better format
s1
```
